# KMeans Clustering

```python
# Elbow Method

from sklearn.cluster import KMeans
wcss = [] #within cluster sum of squares
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init = 'k-means++',random_state = 42)
    kmeans.fit(data1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss, color = 'darkgreen')
plt.title('The Elbow Method',fontsize = 16)
plt.xlabel('Number of Clusters',fontsize = 12)
plt.ylabel('WCSS',fontsize = 12)
plt.show()
```

Optimum number of clusters  = 4

```python
kmeans = KMeans(n_clusters = 4,init = 'k-means++',random_state = 42)
y_kmeans = kmeans.fit_predict(data1)
```

```python
y_kmeans1 = pd.DataFrame(y_kmeans)
```

```python
y_kmeans1[0].value_counts()
```

Data points clustered into 4 clusters

# Agglomerative Hierarchical Clustering

```python
# Dendrogram

import scipy.cluster.hierarchy as sch
plt.figure(figsize = (8,8))
dendrogram = sch.dendrogram(sch.linkage(data1, method = 'ward'))
plt.title('Dendrogram')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean',linkage = 'ward')
y_hc = hc.fit_predict(data1)
```

```python
y_hcfinal = pd.DataFrame(y_hc)
y_hcfinal.shape
```

```python
y_hcfinal[0].value_counts()
```

Data points divided into 2 clusters
